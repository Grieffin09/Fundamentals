```python
# merging and creating the dataframe
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
                 'B':['B0','B1','B2','B3'],
                 'C':['C0','C1','C2','C3'],
                 'D':['D0','D1','D2','D3'],
                 })

df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
                 'B':['B4','B5','B6','B7'],
                 'C':['C4','C5','C6','C7'],
                 'D':['D4','D5','D6','D7'],
                 },index=[1,2,3,4])

df3=pd.DataFrame({'A':['A8','A9','A10','A11'],
                 'B':['B8','B9','B10','B11'],
                 'C':['C8','C9','C10','C11'],
                 'D':['D8','D9','D10','D11'],
                 },index=[8,9,10,11])
```

```python
#Concatenation

df_cat1=pd.concat([df1,df2,df3],axis=0) #axis 0 for the concatenation along column and 1 for row
df_cat1
```

```python
print("Row with assigned index 3 using loc ","\n",df_cat1.loc[3]) #will print the rows with index as 3
print("Row with default index 3 using iloc","\n",df_cat1.iloc[3]) #will print the row with default index as 3
print("Our column A and D is :","\n",df_cat1[['A','D']])
```

```python
#Merging by common key

left=pd.DataFrame({'key':['K0','K1','K2','K3'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']
                 })
right=pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'C':['C4','C5','C6','C7'],
                    'D':['D4','D5','D6','D7']})
```

```python
left
```

```python
right
```

```python
lr_concat=pd.concat([left,right],axis=0)
lr_concat
```

```python
lr_merge=pd.merge(left,right,how='inner',on='key')
lr_merge # the common keys K0,K1,K2 and K3 are merged
```

**what if there is no common merger in the dataframes??**
  * We can use the left_on and right_on parameters in that case (go through the documentation for more)


```python
#usign the join operation
#join considers the indexes of the dataframes
left=pd.DataFrame({
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']},
                    index=['K0','K1','K2','K3'])
right=pd.DataFrame({
                    'C':['C4','C5','C6','C7'],
                    'D':['D4','D5','D6','D7']},
                    index=['K0','K1','K2','K3'])
```

```python
left.join(right,how='outer')
```

```python
right.join(left,how='inner')
```
